```python
import pandas as pd

url = 'https://github.com/dispeter/dsci320/blob/main/dataset.csv?raw=true'
spotify = pd.read_csv(url)
spotify
```

```python
import altair as alt

alt.Chart(spotify).mark_bar().encode(
    alt.X('artists'),
    alt.Y('popularity')
).properties(
    width =174,
    height = 174
)
```
